```python
my_numpy1 = np.linspace(0, 15, 20)
```

```python
my_numpy2 = my_numpy1 ** 3
```

```python
my_numpy1
```

```python
my_numpy2
```

```python
plt.plot(my_numpy1, my_numpy2, "y.-")
```

```python
plt.plot(my_numpy2, my_numpy1, "g--")#çizgi çizgi ve yeşil olsun demek
```

```python
plt.subplot(1, 3, 1) #1 row, 3 column, 1. graph demek. Subplot'a alt grafik denir.
plt.plot(my_numpy1, my_numpy2, "r*")
plt.subplot(1, 3, 2)
plt.plot(my_numpy1, my_numpy2, "b+-")
plt.subplot(1, 3, 3)
plt.plot(my_numpy1, my_numpy2, "c--")
```

```python
#İÇ İÇE GEÇMİŞ GRAFİKLER
my_figure = plt.figure() #figür bu kütüphanenin bir nesnesidir ve grafikler türetir.
my_axes = my_figure.add_axes([0.1, 0.1, 0.5, 0.5]) #burada axes türetiliyor. İlk iki parametre grafiğin başlangıç noktasını, son iki parametre eksenlerin uzunluğunu belirliyor
my_axes.plot(my_numpy1, my_numpy2, "g")
my_axes.set_xlabel("X Data") #figürlerde x-y label komutu değişiyor
my_axes.set_ylabel("Y Data")
my_axes.set_title("Alperen abe")
```

```python
my_figure2 = plt.figure()

my_axes3 = my_figure2.add_axes([0.1, 0.1, 0.9, 0.9]) #aynı figürden başka bir tane daha axes türetiyoruz
my_axes3.plot(my_numpy1, my_numpy2, "g--")
my_axes3.set_xlabel("X Data Large")
my_axes3.set_ylabel("Y Data Large")
my_axes3.set_title("Nested Graphics")

my_axes2 = my_figure2.add_axes([0.23, 0.45, 0.35, 0.35])
my_axes2.plot(my_numpy1, my_numpy2, "g")
my_axes2.set_xlabel("X Data Small")
my_axes2.set_ylabel("Y Data Small")
my_axes2.set_title("Small Graphic")
```

```python
#subplots
(my_fig, my_axe) = plt.subplots()
my_numpy3 = np.linspace(0, 15, 20)
my_numpy4 = my_numpy3 + 1
my_axe.plot(my_numpy1, my_numpy2, color="#FF7531", alpha=1) #color renk kodu, alpha saydamlık oranı
my_axe.plot(my_numpy3, my_numpy4, color="#40E0D0", alpha=1)
```

```python
(new_fig, new_axe) = plt.subplots()
new_axe.plot(my_numpy3, my_numpy3 + 2, color="#C0392B", linewidth=1) #linewidth kalınlığı belirliyor
new_axe.plot(my_numpy3, my_numpy3 + 3, color="#1E8449", linewidth=3, linestyle=":")
new_axe.plot(my_numpy3, my_numpy3 + 4, color="blue", linewidth=3, linestyle="-.")
new_axe.plot(my_numpy3, my_numpy3 + 5, color="purple", linewidth=3, linestyle="--", marker="o") #marker dataları gösteriyor ama sadece o çalışıyor
new_axe.plot(my_numpy3, my_numpy3 + 6, color="black", linewidth=1, linestyle="-", marker="3", markersize=10) #marker'ların boyutunu ayarlıyor
```
